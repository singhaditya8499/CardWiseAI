```python
import requests
from bs4 import BeautifulSoup

def getDetails(link):
    try:
        # Send an HTTP GET request to the URL
        response = requests.get(link)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content of the page using BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')
            
            text_content = soup.get_text()
            text_without_newlines = text_content.replace('\n', '')
            # Print or manipulate the HTML content as needed
            print(text_without_newlines)  # This prints the nicely formatted HTML

            # You can also search for specific elements in the HTML using BeautifulSoup methods
            # For example, to find all links on the page:
            # links = soup.find_all('a')
            
            return text_without_newlines

        else:
            print(f"Failed to retrieve the web page. Status code: {response.status_code}")

    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
credit_card_data['details'] = credit_card_data.Link.apply(getDetails)
```
